## Probabilidade em Python


```python
import numpy as np

dados = np.random.randint(1,6,10000)

evento = dados >= 5

prob = np.mean(evento)

print(prob)
```

Conforme o espaço amostral aumenta a chance de cair uma amostra é dimuniuda.

```python
import numpy as np

N = 1000000

dado1 = np.random.randint(1,7,N)
dado2 = np.random.randint(1,7,N)

print(dado1,dado2)

evento = ((dado1 == 1) & (dado2 == 4)) | ((dado1 == 4) & (dado2 == 1))


probabilidade = np.mean(evento)

print(probabilidade)
```

<h2 style="text-align: center;"> <b> Memória: Configurações Essenciais </b> </h2>



<h4>Define a quantidade de memória usada para cache interno de páginas de dados; quanto maior, menos leituras em disco, mas não deve ultrapassar capacidade real da máquina.</h4>

> shared_buffers = 256MB

<h4>Memória usada por operação para ordenações, hashes e joins; é aplicada por execução e por nó de plano, podendo multiplicar rapidamente em consultas paralelas ou múltiplas conexões.</h4>

> work_mem = 16MB

<h4>Memória usada para operações de manutenção como VACUUM, CREATE INDEX e ALTER TABLE; valores maiores aceleram essas tarefas, mas aumentam consumo momentâneo.</h4>

> maintenance_work_mem = 128MB


<h2 style="text-align: center;"> <b> WAL e Checkpoint: Configurações Fundamentais </b> </h2>



<h4>Define o nível de informação gravada no WAL. "replica" permite crash recovery e replicação física; "minimal" reduz WAL mas limita recursos; "logical" habilita replicação lógica.</h4>

> wal_level = replica

<h4>Determina se o COMMIT espera o WAL ser sincronizado fisicamente no disco antes de confirmar a transação; garante durabilidade quando ativado.</h4>

> synchronous_commit = on

<h4>Define o tempo máximo entre checkpoints automáticos; checkpoints gravam páginas sujas da memória no disco e reduzem o tempo de recovery.</h4>

> checkpoint_timeout = 5min

<h4>Limita o volume máximo de WAL acumulado antes de forçar um checkpoint; valores baixos causam checkpoints frequentes, valores altos aumentam tempo de recuperação.</h4>

> max_wal_size = 1GB

<h4>Define o volume mínimo de WAL mantido para reciclagem de arquivos, evitando criação e remoção constante de segmentos.</h4>

> min_wal_size = 80MB

<h4>Grava a página inteira no WAL na primeira modificação após um checkpoint para evitar corrupção caso ocorra falha durante escrita parcial no disco.</h4>

> full_page_writes = on

<h4>Comprime registros de WAL, especialmente útil quando full_page_writes está ativo; reduz I/O à custa de pequeno uso adicional de CPU.</h4>

> wal_compression = on


<h2 style="text-align: center;"> <b> Logs: Observações </b> </h2>

<h4> Permitir o postgres gravar os logs, porém se estiver no docker precisa mapear o arquivo do docker pro host.  - postgres_log:/var/lib/postgresql/data/log

</h4>

>  loggining_collector = on

<h4>Define onde o postgre envia os logs, se for no docker vai direto pro docker logs container.</h4>

>  log_destination = 'stderr'

<h4> Define pasta onde logs são salvos. </h4>

>  log_directory = 'log'

<h4> Exibe estatísticas detalhadas no WAL no log, útil para labs.</h4>

>  log_wal_status = on

<h4>Permite medir o tempo gasto em escrita no wal.</h4>

> track_wal_io_timing = on

<h4>LDefine padrão do nome do arquivo de log.</h4>

>  log_filename = 'postgresql-%Y-%m-%d_%H%M%S.log'

<h4>
Importante para rastrear as sessões, inclui timestamp, PID, Usuário e database. Sem isso logs são inúteis.    
</h4>


>  log_line_prefix = '%m [%p] %u@%d '

<h4> O postgre pode - registrar quando o checkpoint: começou, duração, nº de páginas escritas e quanto wal consumido.</h4>

>  log_checkpoints = on

<h4>Ajuda a entender mensagens complexas no crash recovery</h4>

> log_error_verbosity = verbose

<h4>Loga todas nova conexão ao banco, bom para segurança mas em produção gera muito logs</h4>

>  log_connections = on

<h4>Registra quando a conexão fecha, consegue medir: tempo de sessão, conexões que morrem rápido demais e se aplicação está usando pool</h4>

>  log_disconnections = on

<h4>Loga quanto tempo cada comando levou, porém pode ter duplicidade se estiver junto de log_min_duration_statement = 0 </h4>

>  log_duration = on

<h4><b>APENAS no lab:</b> loga todas as queries com tempo, podemos cruzar: query pesada, volume de wal e latência de commit.Em produção gera muito log.</h4>

>  log_min_duration_statement = 0

<h4>Vaccum gera wal, com isso você observa: quando autovaccum rodou, quanto tempo durou, quantas páginas limpou<br>update pesado -> + vaccum -> mais wal.</h4>

>  log_autovacuum_min_duration = 0
